```python
df.columns
```

```python
movies = df[['id', 'title', 'genres', 'overview']].rename(columns={'genres': 'genre'})
```

```python
movies = df[['id','title','genres','overview']].rename(columns={'genres':'genre'}).copy()
movies['overview'] = movies['overview'].fillna('')
```

```python
movies
```

```python
movies['tag'] = movies['overview'] + movies['genre']
movies.head(5)
```

```python
new_data = movies.drop(columns=['overview','genre'])
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vector= cv.fit_transform(new_data['tag'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
new_data[new_data['title']=="The Godfather"].index[0]
```

```python
distance= sorted(list(enumerate(similarity[2])), reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index = new_data[new_data['title']==movies].index[0]
    distance= sorted(list(enumerate(similarity[index])), reverse=True,key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
        
```

```python
recommend("Iron Man")
```
